```python
import cv2

cap=cv2.VideoCapture(0)
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

#record top left corner
x=width//2  #// allow us to get value in int
y=height//2

w = width//4
h = height//4

#bottom right x+w,y+h
while True:
    ret, frame=cap.read()
    cv2.rectangle(frame,(x,y),(x+w,y+h),color=(0,0,255),thickness=4)
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

#callback fn rectangle
def draw_rect(event,x,y,flags,param):
    global pt1,pt2,topLeftClick,bottomRightClick
    
    if event==cv2.EVENT_LBUTTONDOWN:
        #reseting the rect
        #it checks if rect there
        if topLeftClick == True and bottomRightClick == True:
            pt1=(0,0)
            pt2=(0,0)
            topLeftClick=False
            bottomRightClick=False
        if topLeftClick == False:
            pt1=(x,y)
            topLeftClick=True
        elif bottomRightClick == False:
            pt2=(x,y)
            bottomRightClick=True
    
#global variable
pt1=(0,0)
pt2=(0,0)
topLeftClick=False
bottomRightClick=False #on click they will update pt1,2

#connect to callback
cap=cv2.VideoCapture(0)
cv2.namedWindow('Test')
cv2.setMouseCallback('Test',draw_rect)

#bottom right x+w,y+h
while True: 
    ret, frame=cap.read()
    #draw frame on the based of global variable
    if topLeftClick:
        #or == True:
        cv2.circle(frame,center=pt1,radius=5,color=(0,0,255),
                  thickness=-1)
        if topLeftClick and bottomRightClick:
            cv2.circle(frame,center=pt2,radius=5,color=(0,0,255),
                  thickness=-1)
            cv2.rectangle(frame,pt1,pt2,(0,0,255),3) #3 is thickness
    
    cv2.imshow('Test', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
#drawing blue circle

import cv2

#callback fn rectangle
def draw_cir(event,x,y,flags,param):
    global center,Click
    
    if event==cv2.EVENT_LBUTTONDOWN:
        #reseting the circle
        #it checks if circle there
        center=(x,y)
        Click=False
    if event == cv2.EVENT_LBUTTONUP:
        Click=True
    
#global variable
center=(0,0)
Click=False

#connect to callback
cap=cv2.VideoCapture(0)
cv2.namedWindow('Test')
cv2.setMouseCallback('Test',draw_cir)

#bottom right x+w,y+h
while True: 
    ret, frame=cap.read()
    #draw frame on the based of global variable
    if Click:
        #or == True:
        cv2.circle(frame,center=center,radius=25,color=(255,0,0),
                  thickness=5)    
    cv2.imshow('Test', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
