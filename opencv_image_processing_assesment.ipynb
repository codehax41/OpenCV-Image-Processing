```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
img=cv2.imread('images/giraffes.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
display_img(img)
```

```python
#applying binary threshold
#Thresholds
img=cv2.imread('images/giraffes.jpg',0)
ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
display_img(th1, cmap='gray')
```

```python
#open angain and change the colorspace to hsv
# Grab V channel
img=cv2.imread('images/giraffes.jpg')
img=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
display_img(img)
```

```python
#create a loss pass filter with 4 by 4 kernel filled with 1/10 and use
#2d convo to blur giraf image and display in normal rgb
kernel=np.ones(shape=(4,4),dtype=np.float32)/10
img=cv2.imread('images/giraffes.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
results=cv2.filter2D(img,-1,kernel)
display_img(results)
```

```python
#create horizontal sobel filter with k=5 to gray scale 
#version of giraf image and display gradient filter version of image
img=cv2.imread('images/giraffes.jpg',0)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
display_img(sobelx, cmap='gray')
```

```python
#plot color hist for r,g,b
img=cv2.imread('images/giraffes.jpg')
show_giraf = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Giraf Image')
plt.show()
```
