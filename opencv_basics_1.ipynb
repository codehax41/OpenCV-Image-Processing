```python
import cv2
import matplotlib.image as mpimg
%matplotlib qt

image=mpimg.imread('images/dog.jpeg')
gray_image=cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image, cmap='gray')
```
